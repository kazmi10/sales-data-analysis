#DATA PREPROCESSING


```python
# Check unique values in the 'rating' column
sales_data['rating'].unique()
```

```python
# Remove any rows with non-numeric ratings
sales_data = sales_data[sales_data['rating'].str.isnumeric()]

# Alternatively, you can replace non-numeric ratings with NaN and then fill them
# sales_data['rating'] = pd.to_numeric(sales_data['rating'], errors='coerce')
# sales_data['rating'].fillna(sales_data['rating'].mean(), inplace=True)

# Now convert the 'rating' column to float
sales_data['rating'] = sales_data['rating'].astype(float)
```

```python
# Verify data types
sales_data.dtypes

# Check for any remaining issues
sales_data.head()
```

```python
# Clean and convert price columns to numeric
sales_data['discounted_price'] = sales_data['discounted_price'].replace('[₹,]', '', regex=True).astype(float)
sales_data['actual_price'] = sales_data['actual_price'].replace('[₹,]', '', regex=True).astype(float)

# Convert discount_percentage to numeric by removing the percentage symbol
sales_data['discount_percentage'] = sales_data['discount_percentage'].replace('[%]', '', regex=True).astype(float)

# Convert rating and rating_count to numeric
sales_data['rating'] = pd.to_numeric(sales_data['rating'], errors='coerce')
sales_data['rating_count'] = pd.to_numeric(sales_data['rating_count'].replace('[,]', '', regex=True), errors='coerce')

# Verify the changes
sales_data.dtypes
```

```python
# Check for missing values
sales_data.isnull().sum()

# Optionally fill or drop missing values
# Example: Fill missing values in 'rating_count' with the mean of the column
sales_data['rating_count'].fillna(sales_data['rating_count'].mean(), inplace=True)

# You could also drop rows with missing values if that's more appropriate
# sales_data.dropna(inplace=True)
```

```python
# Drop irrelevant columns if not needed
sales_data.drop(['img_link', 'product_link'], axis=1, inplace=True)

# Verify the cleaned dataset
sales_data.head()
```

#Customer Segmentation

```python
# Group data by user_id to get customer-level data
customer_data = sales_data.groupby('user_id').agg({
    'discounted_price': 'sum',       # Total amount spent
    'product_id': 'count',           # Number of purchases
    'discount_percentage': 'mean',   # Average discount received
}).rename(columns={'discounted_price': 'total_spent',
                   'product_id': 'num_purchases',
                   'discount_percentage': 'avg_discount'})

# Display the first few rows of the summarized data
customer_data.head()
```

```python
from sklearn.preprocessing import StandardScaler

# Standardizing the features
scaler = StandardScaler()
customer_data_scaled = scaler.fit_transform(customer_data)

# Converting the scaled data back to a DataFrame
customer_data_scaled = pd.DataFrame(customer_data_scaled, columns=customer_data.columns)

# Display the first few rows of the scaled data
customer_data_scaled.head()
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Calculate the sum of squared distances (inertia) for a range of cluster numbers
inertia = []
K = range(1, 11)

for k in K:
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(customer_data_scaled)
    inertia.append(kmeans.inertia_)

# Plot the Elbow curve
plt.figure(figsize=(8, 4))
plt.plot(K, inertia, 'bo-')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal Number of Clusters')
plt.show()
```

```python
# Let's assume the optimal number of clusters is 3 based on the elbow method
optimal_clusters = 3

# Applying K-Means
kmeans = KMeans(n_clusters=optimal_clusters, random_state=0)
customer_data['Cluster'] = kmeans.fit_predict(customer_data_scaled)

# Display the cluster centers and the first few rows of clustered data
print(kmeans.cluster_centers_)
customer_data.head()
```

```python
# Group by cluster and calculate mean values for each cluster
cluster_analysis = customer_data.groupby('Cluster').mean()

# Display the cluster analysis
cluster_analysis
```

#Analysis of Customer Segmentation

##Based on the results from the K-Means clustering:

###Elbow Method Result:

The Elbow Method suggests that 3 clusters may be a good choice, as the curve starts to bend at this point. This implies that adding more clusters beyond 3 does not significantly reduce the within-cluster variance (inertia).

###Cluster Centers:
The cluster centers provided indicate the average values of the features (total_spent, num_purchases, avg_discount) for each cluster. Here's what each center represents:
Cluster 0: Customers who spend an average of 1294.77, make 1 purchase, and receive an average discount of around 50.91%.
Cluster 1: Customers who spend an average of 1755.94, make 1 purchase, and receive an average discount of 21.12%.
Cluster 2: Customers who spend a significantly higher amount, averaging 9469.00, make 1 purchase, and receive an average discount of 37.60%.

###Cluster Characteristics:
Cluster 0: Represents budget-conscious customers who tend to purchase with high discounts.
Cluster 1: These are mid-range customers who spend a moderate amount but receive lower discounts. They might be less price-sensitive.
Cluster 2: This cluster represents high-value customers who spend significantly more and receive a moderate discount. These could be premium customers who prioritize product quality or brand over discounts.

##Actionable Insights:
Targeting Cluster 0: You could create marketing campaigns focused on offering even better discounts or bundled offers to these customers to increase their purchase frequency.

Targeting Cluster 1: These customers might respond well to loyalty programs or offers that emphasize value rather than just discounts. They might be encouraged to make repeat purchases through exclusive membership benefits.

Targeting Cluster 2: Given their high spending, premium services, personalized offers, and exclusive access to new or high-end products could be effective in retaining these customers and increasing their lifetime value.
